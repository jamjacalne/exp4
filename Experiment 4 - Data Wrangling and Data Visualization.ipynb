## Data Wrangling

```python
import pandas as pd
```

```python
x = pd.read_csv('board2.csv')
x
```

```python
vis = x[(x['Hometown'] == 'Visayas') & (x['Math'] > 70)][['Name', 'Gender', 'Track', 'Math']].reset_index(drop=True)
vis
```

```python
instru = x[(x['Hometown'] == 'Luzon') & (x['Track'] == 'Instrumentation') & (x['Electronics'] > 70)][['Name', 'GEAS', 'Electronics']].reset_index(drop=True)
instru
```

```python
x['Average'] = x[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)
mindy = x[(x['Hometown'] == 'Mindanao') & (x['Gender'] == 'Female') & (x['Average'] >= 55)][['Name', 'Track', 'Electronics', 'Average']].reset_index(drop=True)
mindy
```

## Data Visualization: Average Grade according to a. Hometown, b. Gender, c. Hometown

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(x='Track', y='Average', data=x, estimator='mean')
plt.title('Average Grade by Track')
plt.show()
```

```python
sns.barplot(x='Gender', y='Average', data=x, estimator='mean')
plt.title('Average Grade by Gender')
plt.show()
```

```python
sns.barplot(x='Hometown', y='Average', data=x, estimator='mean')
plt.title('Average Grade by Hometown')
plt.show()
```
